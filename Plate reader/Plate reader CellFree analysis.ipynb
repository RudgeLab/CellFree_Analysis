```python
#dimensionando los datos:
#in the cell mode of datasheet index start from 1 and no from 0 as usually on python
rows=1 #start from the second row
columns=1 #and the third row column  ---> this is to avoid problems with headers
while True:
    if wsC.cell(row=rows+1, column=columns+1).value != None:
        rows+=1
    else:
        break
while True:
    if wsC.cell(row=2, column=columns+1).value != None:
        columns+=1
    else:
        break
print(str(rows) + ' rows') 
print(str(columns) + ' columns')
```

```python
dataC = np.zeros((rows-2,columns-2)) #in this case, the first two rows are non numeric values
timeC = np.zeros((rows-2))
headers = []
add = True
for i in range(rows-2):
    for j in range(columns-2):
        dataC[i,j]=wsC.cell(row=i+3, column=j+3).value
        if add:
            headers.append(wsC.cell(row=2, column=j+3).value)  # list with the headers
    add = False
for i in range(rows-2):
    time_aux = wsC.cell(row=i+3, column=1).value
    #print(time_aux)
    timeC[i] =time_aux.hour + time_aux.minute/60 + time_aux.second/3600
    print(timeC[i])
print(headers)
```

```python
#define the blank data series
B_rows_lab=['D1','D2','D3','D5','D6','D7','E1','E2','E3','E5','E6','E7']  #fill this list manually
B_cols= []
for name in B_rows_lab:
    if name in headers:
        #print(name)
        B_cols.append(headers.index(name))
print(B_cols)
```

```python
#compute the mean blank values for each time
B_cols_valuesC=np.zeros(((rows-2),len(B_cols)))
#print(B_cols_valuesC.shape)
for i in range(len(B_cols)):
    B_cols_valuesC[:,i]=dataC[:,B_cols[i]]

B_MvalsC=np.mean(B_cols_valuesC, axis=1)

B_Mvals_MC=np.zeros(dataC.shape)

for i in range(dataC.shape[1]):
    B_Mvals_MC[:,i]=B_MvalsC[:]     #this is to perform the substraction over the data

fig= plt.figure()
fig.suptitle('Blank mean serie')
plt.plot(timeC, B_MvalsC)
plt.ylabel('Intensity [AU]')
plt.xlabel('Time [hr]')
plt.show()

print(B_MvalsC.shape)

fig= plt.figure()
fig.suptitle('Blanks series')
plt.plot(timeC, dataC[:,B_cols])
plt.ylabel('Intensity [AU]')
plt.xlabel('Time [hr]')
plt.show()
```

```python
#compute the mean blank values for each time
B_cols_valuesY=np.zeros(((rows-2),len(B_cols)))
#print(B_cols_valuesC.shape)
for i in range(len(B_cols)):
    B_cols_valuesY[:,i]=dataY[:,B_cols[i]]

B_MvalsY=np.mean(B_cols_valuesY, axis=1)

B_Mvals_MY=np.zeros(dataC.shape)

for i in range(dataY.shape[1]):
    B_Mvals_MY[:,i]=B_MvalsY[:]     #this is to perform the substraction over the data

fig= plt.figure()
fig.suptitle('Blank mean serie')
plt.plot(timeY, B_MvalsY)
plt.ylabel('Intensity [AU]')
plt.xlabel('Time [hr]')
plt.show()

print(B_MvalsC.shape)

fig= plt.figure()
fig.suptitle('Blanks series')
plt.plot(timeY, dataY[:,B_cols])
plt.ylabel('Intensity [AU]')
plt.xlabel('Time [hr]')
plt.show()
```

```python
#substract blank value from the rest of the data

#Yellow
dataBY=np.zeros(dataY.shape)
dataBY=dataY-B_Mvals_MY

#Cyan
dataBC=np.zeros(dataC.shape)
dataBC=dataC-B_Mvals_MC
```

```python
#don't allow negative values
for i in range(dataC.shape[0]):
    for j in range(dataC.shape[1]):
        if dataBY[i,j] <= 0 :
            dataBY[i,j] = 0.0001
        
        if dataBC[i,j] <= 0 :
            dataBC[i,j] = 0.0001
```

```python
#define the interest data series
I_rows_lab=['A1','A2','A3','A5','A6','A7','B1','B2','B3','B5','B6','B7','C1','C2','C3','C5','C6','C7']  #fill this list manually
I_cols= []
for name in I_rows_lab:
    if name in headers:
        #print(name)
        I_cols.append(headers.index(name))
print(I_cols)
```

```python
ratio=Yellow/Cyan
print(ratio)
print(ratio.shape)
```

```python
plt.plot(timeC[20:60], ratio[20:60,:])
plt.ylabel('Intensity [AU]')
plt.xlabel('Time [hr]')
plt.show()
```

```python
#perform linear fit
z=[]
init=25    #init of the "linear" region
end=90    #end of the "linear region
xp = np.linspace(timeC[init], timeC[end], 1000)
for i in range(ratio.shape[1]):
    z.append(np.polyfit(timeC[init:end], ratio[init:end,i], 1))
    print(z[-1])           #first component is the slope
    p = np.poly1d(z[-1])
    #plt.plot(timeC[init:end], ratio[init:end,i], '.', xp, p(xp), '-')
    plt.plot(timeC[14:-1], ratio[14:-1,i], '.', xp, p(xp), '-')
    plt.show()
    
    
```

```python
ratiometric = np.zeros(len(z))
for i in range(len(z)):
    ratiometric[i]=z[i][0]
print(ratiometric)
```

```python
Mean_R=np.mean(ratiometric)
StdDesv=np.std(ratiometric)
print(Mean_R)
print(StdDesv)
```

```python
Ratiometric2=np.zeros(len(ratiometric)-3)
counter=0
for i in range(len(ratiometric)):
    if i != 3 and i != 4 and i !=5:
        Ratiometric2[counter]=ratiometric[i]
        counter+=1
print(Ratiometric2.shape)
```

```python
Mean_R=np.mean(Ratiometric2)
StdDesv=np.std(Ratiometric2)
print(Mean_R)
print(StdDesv)
print(Ratiometric2)
```

```python
fig= plt.figure()
plt.plot(Ratiometric2,'r.')
fig.suptitle('Ratiometric')
plt.ylabel('Intensity [AU]')
plt.xlabel('Time [hr]')
plt.show()
fig.savefig('Ratiometric.tif')
```
